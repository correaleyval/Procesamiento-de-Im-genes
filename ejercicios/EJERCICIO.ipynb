# Ejercicio 2

```python
lena = imread('lena_gray_512.tif')
```

```python
imshow(lena)
```

```python
mask = zeros_like(lena)
```

```python
from skimage.draw import circle
```

```python
cx = int(lena.shape[0]/2)
cy = int(lena.shape[1]/2)
rr, cc = circle(cx, cy, 150, shape=lena.shape)
```

```python
mask[rr, cc] = 1
```

```python
imshow(mask)
```

```python
R = lena*mask
```

```python
imshow(R)
```
